## CUSTOMER 
PRAGMA foreign_keys = 0;

CREATE TABLE sqlitestudio_temp_table AS SELECT *
                                          FROM customer;

DROP TABLE customer;

CREATE TABLE customer (
    ID                    INTEGER PRIMARY KEY
                                  UNIQUE,
    name                  STRING,
    age                   INTEGER,
    address               TEXT,
    [favourite artist]    STRING,
    [money spent on art ] REAL
);

INSERT INTO customer (
                         ID,
                         name,
                         age,
                         address,
                         [favourite artist],
                         [money spent on art ]
                     )
                     SELECT ID,
                            name,
                            age,
                            address,
                            "favourite artist",
                            "money spent on art "
                       FROM sqlitestudio_temp_table;

DROP TABLE sqlitestudio_temp_table;

PRAGMA foreign_keys = 1;
